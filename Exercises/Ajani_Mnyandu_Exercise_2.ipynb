```python
# Import modules

import math
import matplotlib.pyplot as plt
```

```python
# Define variables

time = range(0, 21, 1) # Days
k_prime = 0.1 # 1/day
p_naught = 0.1 # mmol N m^-3, phytoplankton initial population
n_naught = 5 # mmol N m^-3
m = 5 # Asymptote
```

```python
# Set up the equations

x = p_naught + n_naught
n = [n_naught * math.exp(-1*k_prime*x*t) for t in time]
p_t = [p_naught*(x/(p_naught+i)) for i in n]
n_t = [x - p for p in p_t]
a = [m]*len(time)
```

```python
# Define plot figures

title = 'Time evolution of phytoplankton growth and nutrient concentration'
xlabel = 'time (days)'
ylabel1 = 'Phytoplankton growth (mmol N /m$^3$)'
ylabel2 = 'Nutrient concentration (mmol N /m$^3$)'
label1 ='Nutrient concentration'
label2 = 'Phytoplankton growth'
```

```python
# Plot the graph

fig, ax1 = plt.subplots()

ax1.plot(time, n_t, color = 'b', label =label1)
ax1.plot(time, a, linestyle = 'dotted', color = 'black')
ax1.set_xlabel(xlabel)
ax1.set_ylabel(ylabel2)
plt.legend(loc = 'center right')
plt.grid(True)

ax2 = ax1.twinx()

ax2.plot(time,p_t,linestyle = '--', color= 'g', label =label2)
ax2.set_ylabel(ylabel1)
plt.legend(loc = [0.55, 0.55])

plt.suptitle(title)
plt.show()
```
