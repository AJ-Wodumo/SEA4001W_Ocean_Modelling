```python
# import modules

import pylab as pl
```

```python
# define variables

t0 = 0.0  # time is in days
tn = 20.0
dt_values = [1,2,5,10]  # different values of ∆t
kappa = 0.05  # day^−1
C0 = 0.5  # mg chl/m^3
```

```python
# Define the right hand side of the ODE

def F(k, C):
    return -k * C # The right hand section of the equation
```

```python
# Define the euler backward method

def euler_backward(dt):
    Ntot = int(pl.floor((tn - t0) / dt)) + 1
    C = pl.zeros(Ntot)
    t = pl.zeros(Ntot)
    t[0] = t0
    C[0] = C0

    for n in range(Ntot - 1):
        C[n + 1] = C[n] / (1 + kappa * dt)
        t[n + 1] = t[n] + dt

    return t, C
```

```python
# Define plot features

Xdim = 10
Ydim = 6
xlabel = 'Time (days)'
ylabel = 'Concentration (mg chl/m^3)'
title = 'Euler-Backward Method for different time steps'
```

```python
# Plot the figure

pl.figure(figsize=(Xdim, Ydim))

for dt in dt_values: # Generate different euler backward method graphs
    t, C = euler_backward(dt)
    pl.plot(t, C, linestyle = 'dotted',label=f'dt = {dt}')

pl.xlabel(xlabel)
pl.ylabel(ylabel)
pl.title(title)
pl.legend()
pl.grid(True)
pl.show()
```

#### At large values of ∆t, the more the graphs diverge at increasingly large time values.
