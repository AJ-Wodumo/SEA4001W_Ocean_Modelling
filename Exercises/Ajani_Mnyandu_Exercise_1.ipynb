```python
# Import modules

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data

growth_data = pd.read_csv('../Data/flynn94.csv')
print(growth_data.head())
growth_data.keys()
```

```python
# Read dataframe columns

time = growth_data['time']
cells = growth_data['cells']
biovolume = growth_data['biovolume']
ammonium = growth_data['ammonium']
cellC = growth_data['cellC']
chl_a = growth_data['chl-a']
```

```python
# Define plot features

title = 'Growth of phytoplankton vs time'
xlabel = 'time (days)'
ylabel1 = 'cell number (10$^6$ ml$^-1$)'
ylabel2 = 'biovolume (nl/ml)'
ylabel3 = 'Ammonium ($10^-6$ g N ml$^-1$)'
ylabel4 = 'CellC ($10^-6$ C N ml$^-1$)'
ylabel5 = 'chl-a (ng ml$^-1$)'
xdim  =  15
ydim = 10
```

```python
# Define plot data

data = [cells, biovolume, ammonium, cellC, chl_a]
y_labels = [ylabel1, ylabel2, ylabel3, ylabel4, ylabel5]
variables = ['Cell Number', 'Biovolume', 'Ammonium', 'Cellular Carbon', 'Chlorophyll']
```

```python
# Generate the plots

fig, ax = plt.subplots(3, 2, figsize = (xdim, ydim))
ax = ax.flatten()

for i in range(len(data)):
    ax[i].plot(time, data[i])
    ax[i].set_xlabel(xlabel)
    ax[i].set_ylabel(y_labels[i])
    ax[i].set_title(f'Phytoplankton growth in {variables[i]} vs time')

ax[-1].axis('off')
plt.suptitle(title, fontsize = 20)
plt.tight_layout()
plt.show()
```

## Questions

### 1. Is the growth exponential?

Yes. The growth is exponential. Between day 5 and day 18, the rate of cell growth, across all of the markers (except ammonium) accelerates.

### 2. Can you separate the growth curve into some phases?

 Yes. There are 4 distinct phases that can be seen in the curve:
1. The initial phase occurs between day 0 - 5. The cell population is small, and rate of cell division is low as the cells adapt to the environmental conditions.
2. The exponential phase occurs between day 6 to 18. The rate of cell division accelerates, which increases the phytoplankton growth.
3. Between 12 and 18, the rate of cell division begins to slow down as all of the nutrients have been depleted.
4. The stationary phase occurs between day 19 - 25. The rate of cell division is very small, and there are no major increases in the population of the phytoplankton.

### 3. How different is the growth expressed in cellular carbon from the one in cell number and chlorophyll?

The exponential phase of the cellular carbon is more gentle when compared to the one in cell number and chlorophyll (chl-a) and it might indicate that the phytoplankton are still undergoing cell division despite there being no change in the population of the phytoplankton beyond day 18, and the depletion of ammonium in the tank. 

### 4. Is the ammonium consumption approximated by an exponential curve?

Yes, the ammonium consumption is approximated by an exponential curve, as the ammonium sharply declines after day 7 before being completely depleted by day 12.
